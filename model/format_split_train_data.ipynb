```python
def strip_input(input,si,ei):
    print(f"got si={si}, got ei={ei}")
    if type(si) is not float or np.isnan(si) or si<0 :
        si=0
    if type(ei) is not float or np.isnan(ei) or ei<0 or ei>len(input):
        si=len(input)-1
    si=int(si)
    ei=int(ei)
    return input[si:ei]
```
